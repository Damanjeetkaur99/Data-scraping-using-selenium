```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import pandas as pd
from csv import DictWriter
import openpyxl
from openpyxl import load_workbook
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')


driver = webdriver.Chrome(executable_path="/Users/apple/Downloads/chromedriver3")
url="http://sec2021.bihar.gov.in/claim-objection/Final_WinningList.aspx"
driver.get(url)
driver.maximize_window()
sleep(5)


post = '//*[@id="ctl00_ContentPlaceHolder1_ddl_Post"]'
driver.find_element(by=By.XPATH, value=(post)).click()
sleep(2)
post_counts = len(driver.find_elements(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_ddl_Post"]/option'))
print("post", post_counts)
for i in range(2,post_counts):
    driver.find_element(by=By.XPATH, value=(post +'/option['+str(i)+']')).click()
    sleep(2)
    
    dist = '//*[@id="ctl00_ContentPlaceHolder1_ddlDistrict"]'
    driver.find_element(by=By.XPATH, value=(dist)).click()
    sleep(2)
    dist_counts = len(driver.find_elements(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_ddlDistrict"]/option'))
    print("district", dist_counts)
    for j in range(2,40):
        driver.find_element(by=By.XPATH, value=(dist + '/option[' + str(j) + ']')).click()
        sleep(1)
        
        block = '//*[@id="ctl00_ContentPlaceHolder1_ddlBlock"]'
        driver.find_element(by=By.XPATH, value=(block)).click()
        sleep(1)
        block_counts = len(driver.find_elements(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_ddlBlock"]/option'))
        print("block", block_counts)
        df = pd.DataFrame()
        for k in range(2,20):
            driver.find_element(by=By.XPATH, value=(block + '/option[' + str(k) + ']')).click()
            sleep(1)
            
            panch = '//*[@id="ctl00_ContentPlaceHolder1_ddlPanchayat"]'
            driver.find_element(by=By.XPATH, value=(panch)).click()
            sleep(2)
            panch_counts = len(driver.find_elements(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_ddlPanchayat"]/option'))
            print("panchayat", panch_counts)
            for p in range(2,22):
                driver.find_element(by=By.XPATH, value=(panch + '/option[' + str(p) + ']')).click()
                sleep(1)

            
                driver.find_element(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_ctl00"]/div[1]/div[5]')).click()
                sleep(1)
                
                
                Post = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[2]'))
                District = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[3]'))
                Block = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[4]'))
                Panchayat = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[5]'))
                Reservation = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[7]'))
                Candidate_Name = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[8]'))
                Father_Husband_Name = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[9]'))
                Gender = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[10]'))
                Age = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[11]'))
                Category = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[12]'))
                Address = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[13]'))
                Phone_No = driver.find_elements(by=By.XPATH, value=('//*[@id="ctl00_ContentPlaceHolder1_GridView1"]/tbody/tr/td[14]'))
                
                sleep(2)
                result = []
                for r in range(len(Post)):
                    data = {'Post': Post[r].text,
                            'District': District[r].text,
                            'Block': Block[r].text,
                            'Panchayat': Panchayat[r].text,
                            'Reservation': Reservation[r].text,
                            'Candidate_Name': Candidate_Name[r].text,
                            'Father_Husband_Name.': Father_Husband_Name[r].text,
                            'Gender': Gender[r].text,
                            'Age': Age[r].text,
                           'Category': Category[r].text,
                           'Address': Address[r].text,
                           'Phone_No': Phone_No[r].text}
                    result.append(data)
                sleep(3)
                df_data = pd.DataFrame(result)
                df = df.append(df_data)
                print(df)
    
    driver.close()            
#             df1 = pd.read_excel('bihar_win_list.xlsx')
#             df1 = df1.append(df)
#             df1.to_excel('bihar_win_list.xlsx', index=False)
```
